### 2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Convert the "orders_day_of_week" column to datetime format
ords_prods_merged['orders_day_of_week'] = pd.to_datetime(ords_prods_merged['orders_day_of_week'])
```

```python
# Calculate the daily order count for each day of the week

daily_order_count = ords_prods_merged['ords_prods_merged'].dt.day_name().value_counts()
```

```python
# Sort the days based on the order count to get the two busiest and two slowest days

busiest_days = daily_order_count.sort_values(ascending=False).head(2).index
slowest_days = daily_order_count.sort_values().head(2).index
```

```python
# Create a new column "busiest_days" and fill with appropriate labels

ords_prods_merged['busiest_days'] = ords_prods_merged['orders_day_of_week'].dt.day_name().apply(lambda x: 'Busiest days' if x in busiest_days else 'Slowest days')
```

### 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Convert the "order_hour_of_day" column to datetime format

ords_prods_merged['order_hour_of_day'] = pd.to_datetime(ords_prods_merged['order_hour_of_day'])
```

```python
# Calculate the hourly order count for each hour of the day

hourly_order_count = ords_prods_merged['order_hour_of_day'].dt.hour.value_counts()
```

```python
# Determine the thresholds for busiest, average, and fewest orders

most_orders_threshold = hourly_order_count.quantile(0.75)
fewest_orders_threshold = hourly_order_count.quantile(0.25)
```

```python
# Create a function to assign labels based on the order count

def label_busiest_period(order_count):
    if order_count >= most_orders_threshold:
        return 'Most orders'
    elif order_count >= fewest_orders_threshold:
        return 'Average orders'
    else:
        return 'Fewest orders'
```

```python
# Create the new column "busiest_period_of_day" using the apply function

ords_prods_merged['busiest_period_of_day'] = ords_prods_merged['order_hour_of_day'].dt.hour.apply(lambda x: label_busiest_period(hourly_order_count[x]))
```
